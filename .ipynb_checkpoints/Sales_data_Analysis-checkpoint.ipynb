```python
import pandas as pd
import glob

path = r'C:\Users\ROHIT\Desktop\python project\Sales data analysis\SalesAnalysis\Sales_Data' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename)
    li.append(df)

sales_data = pd.concat(li, axis=0)
```

```python
sales_data
```

```python
sales_data.isnull().sum()
```

```python
sales_data = sales_data.dropna()
```

```python
sales_data.isnull().sum()
```

```python
sales_data
```

# Adding monthand time columns

```python
sales_data = sales_data[sales_data['Order Date'].str[0:2]!='Or']
```

```python
sales_data
```

```python
sales_data['Month'] = pd.to_datetime(sales_data['Order Date']).dt.month
```

```python
sales_data
```

```python
sales_data['time'] = pd.to_datetime(sales_data['Order Date']).dt.time
```

```python
sales_data
```

# Adding city columns

```python
sales_data['city'] = sales_data['Purchase Address'].apply(lambda x:"".join( x.split(",")[1:3]))
```

```python
sales_data
```

# Question 1: What was the best month for sales? How much was earned that month?

```python
sales_data['Revenue'] = (sales_data['Price Each'].astype('float'))*(sales_data['Quantity Ordered'].astype('int'))
```

```python
best_month = sales_data.groupby('Month').sum()['Revenue'].reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python

sns.barplot(best_month['Month'],best_month['Revenue'])
```
